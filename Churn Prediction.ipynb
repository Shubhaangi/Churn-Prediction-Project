# Hyperparameter Tuning And Model Training

```python
drop_list=['churn']
X = df.drop(drop_list, axis=1)
y = df.churn.values
X.head(5)
```

```python
#splitting the dataset in test set and train set
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
y_train.shape
```

For choosing the correct model for model training, I'll choose some of most common used model in classification problem as this project is classification based project.Also i'll use Recursive Feature Elimination (RFE) via pipeline and fine tune models via grid search.Stratified k fold cross validation will be used because the data was highly imbalanced and for the same reason i will be using ROC AUC score as well.   

```python
#importing all the libraries needed
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_curve,roc_auc_score
from sklearn.metrics import auc, confusion_matrix
from sklearn.decomposition import PCA
from sklearn.model_selection import cross_validate
from sklearn.model_selection import GridSearchCV, cross_val_score, StratifiedKFold, learning_curve
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from  sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier,GradientBoostingClassifier
from sklearn.neighbors import  KNeighborsClassifier   
from sklearn.metrics import accuracy_score,precision_score,recall_score 
from sklearn.pipeline import Pipeline
import seaborn as sns
from sklearn.feature_selection import RFE
from sklearn import feature_selection
```

```python
#Using recursive feature elimination (RFE) via Pipeline 

class PipelineRFE(Pipeline):
    def fit(self, X, y=None, **fit_params):
        super(PipelineRFE, self).fit(X, y, **fit_params)
        self.coef_ = self.steps[-1][-1].coef_
        return self
    
class PipeRFE(Pipeline):

    def fit(self, X, y=None, **fit_params):
        super(PipeRFE, self).fit(X, y, **fit_params)
        self.feature_importances_ = self.steps[-1][-1].feature_importances_
        return self    

clf_knn=KNeighborsClassifier()


pipe_lr = PipelineRFE([('sc', StandardScaler()),('lr',LogisticRegression(random_state=42))])

pipe_svm= PipelineRFE([('sc', StandardScaler()),('svm',SVC(probability=True,random_state=42))])

pipe_dt= PipeRFE([('sc', StandardScaler()),('dt',DecisionTreeClassifier(criterion='entropy',random_state=42))])

pipe_rf= PipeRFE([('sc', StandardScaler()),('rf',RandomForestClassifier(criterion='entropy'))])

pipe_knn =Pipeline([('sc', StandardScaler()),('clf',clf_knn)])

pipe_adb=PipeRFE([('sc', StandardScaler()),('adb',AdaBoostClassifier(random_state=4))])

pipe_gb =PipeRFE([('sc', StandardScaler()),('gb',GradientBoostingClassifier(random_state=42))])

selector_lr = feature_selection.RFE(pipe_lr)
selector_svm = feature_selection.RFE(pipe_svm)
selector_dt = feature_selection.RFE(pipe_dt)
selector_rf = feature_selection.RFE(pipe_rf)
selector_adb = feature_selection.RFE(pipe_adb)
selector_gb = feature_selection.RFE(pipe_gb)

clf_labels = ['Logistic Regression','SVM', 'Decision Tree','Random Forest','KNN','AdaBoost','Gradient Boosting']
all_clf = [selector_lr,selector_svm,selector_dt,selector_rf,pipe_knn,selector_adb,selector_gb]
 
params_lr = [{'estimator__lr__C': [ 0.1, 0.01,10.0,100.0]}]
params_svm=[{'estimator__svm__kernel':['linear'],
           'estimator__svm__C': [ 0.1,1.0,10.0],'estimator__svm__gamma': [ 0.1]}]

params_dt=[{'estimator__dt__max_depth': [5, 6,8]}]
params_rf=[{'estimator__rf__n_estimators':[5,6,7,8,9,10],
           'estimator__rf__random_state':[1,42],'estimator__rf__n_jobs':[-1,2,3,4]}]
params_knn=[{'clf__n_neighbors':[3,4,5,6,8,15],'clf__p':[2],
           'clf__metric':['minkowski']}]
params_adb = [{'estimator__adb__n_estimators' : [50,100,200,300],
              'estimator__adb__learning_rate': [0.1,1.0, 0.05, 0.01]}]
params_gb = [{'estimator__gb__loss' : ["deviance"],
              'estimator__gb__n_estimators' : [100,200,300],
              'estimator__gb__learning_rate': [0.1, 0.05, 0.01],
              'estimator__gb__max_depth': [4, 8],
              'estimator__gb__min_samples_leaf': [100,150],
              'estimator__gb__max_features': [0.3, 0.1] 
              }]


cv1=StratifiedKFold(n_splits=5, random_state=None)
#Initiating Grid Search
grid_lr = GridSearchCV(estimator=selector_lr, param_grid=params_lr,cv=cv1,scoring='roc_auc') 
 
grid_svm= GridSearchCV(estimator=selector_svm, param_grid=params_svm,cv=cv1,scoring='roc_auc')

grid_dt= GridSearchCV(estimator=selector_dt, param_grid=params_dt,cv=cv1,scoring='roc_auc')
 
grid_rf = GridSearchCV(estimator=selector_rf, param_grid=params_rf,cv=cv1,scoring='roc_auc')

grid_knn= GridSearchCV(estimator=pipe_knn, param_grid=params_knn,cv=cv1,scoring='roc_auc')

grid_adb= GridSearchCV(estimator=selector_adb, param_grid=params_adb,cv=cv1,scoring='roc_auc')

grid_gb= GridSearchCV(estimator=selector_gb, param_grid=params_gb,cv=cv1,scoring='roc_auc')

grid=[grid_lr,grid_svm,grid_dt,grid_rf,grid_knn,grid_adb,grid_gb]
grid_dict={0:'Logistic Regression',1:'SVM',2:'Decision Tree',3:'Random Forest',4:'KNN',5:'AdaBoost',6:'Gradient Boosting'}
print('Model Optimization...')
best_score=0.0
best__clf=0
best_gs=''
for idx,gs in enumerate(grid):
    print('Estimator:%s'%grid_dict[idx])
    gs=gs.fit(X_train,y_train)
    print('Best parameters: %s'%gs.best_params_)
    print('Best training roc auc score:%.2f'%gs.best_score_)
    y_pred=gs.predict(X_test)
    print('Test dataset roc auc score for best parameteres:%.2f'%roc_auc_score(y_test,y_pred))
    if roc_auc_score(y_test,y_pred)>best_score:
        best_score=roc_auc_score(y_test,y_pred)
        best_gs=gs
        best_clf=idx
print('\nClassifier with best test set ROC AUC Score:%s'%grid_dict[best_clf])
    
```

So after training our model on training set and then testing it on test set ,the conclusion drawn is that Decision tree is the best estimator.Decision Tree is also the best choice as it only shows a slight degree of overfitting in comparison to other estimators. Now lets evaluate our result.

Now lets take a look at what the probability of churned customer will be through ROC curve.

```python
from scipy import interp
def plot_roc_auc_curve(y_test,estimator,title,pos_label=1,njobs=-1):
    fig = plt.figure(figsize=(7, 5))
    tprs = []
    mean_fpr = np.linspace(0, 1, 100)
    all_tpr = []
    aucs = []
    fpr, tpr, thresholds = roc_curve(y_test,estimator)
    tprs.append(interp(mean_fpr, fpr, tpr))
    tprs[-1][0] = 0.0
    roc_auc = auc(fpr, tpr)
    aucs.append(roc_auc)
    plt.plot(fpr,tpr,lw=1,label='ROC Curve(area = %0.2f)'%( roc_auc))
    plt.plot([0, 1],[0, 1],linestyle='--',color='g',label='random guessing')
    plt.plot([0, 0, 1],[0, 1, 1],lw=2,linestyle=':',color='r',label='perfect performance')
    plt.xlim([-0.05, 1.05])
    plt.ylim([-0.05, 1.05])
    plt.xlabel('false positive rate')
    plt.ylabel('true positive rate')
    plt.title(title)
    plt.legend(loc="lower right")
    return plt

plot_roc_auc_curve(y_test,y_prob_lr,"Logistic Regression ROC curves")
plot_roc_auc_curve(y_test,y_prob_svm,"SVM ROC curves")
plot_roc_auc_curve(y_test,y_prob_dt,"Decision Tree ROC curves")
plot_roc_auc_curve(y_test,y_prob_rf,"Random Forest ROC curves")
plot_roc_auc_curve(y_test,y_prob_knn,"K Nearest Neighbor ROC curves")
plot_roc_auc_curve(y_test,y_prob_adb,"Ada Boost ROC curves")
plot_roc_auc_curve(y_test,y_prob_gb,"Gradient Boosting ROC curves")
plt.show()
```

Well from above graphs it can be seen that Gradient Boosting gives highest probability for predicting churned customer.And Decision tree rank second to gradient boosting with a slight difference of 0.01. But from the prediction result through ROC AUC score, Decision tree gave highest test score. ## Confusion Matrix

```python


class_names = ['Not Churn','Churn']   

def plot_confusion_matrix(cm,title,classes,normalize=False,
                          cmap=plt.cm.Blues):
        
        
        plt.imshow(cm,interpolation='nearest',cmap=plt.cm.Blues)
        plt.title(title)
        plt.colorbar()
        tick_marks = np.arange(len(classes))
        plt.xticks(tick_marks, classes, rotation=45)
        plt.yticks(tick_marks, classes)
        fmt = 'd'
        for i in range(cm.shape[0]):
            for j in range(cm.shape[1]):
                plt.text(x=j,y=i,s=format(cm[i, j], fmt),va='center',ha='center')
        plt.xlabel('Predicted')
        plt.ylabel('True')
        plt.tight_layout()
        plt.figure()
        return plt
    
 
   

plot_confusion_matrix(confusion_matrix(y_test, y_pred_lr),"linear regression",classes=class_names)
plot_confusion_matrix(confusion_matrix(y_test, y_pred_svm),"Support Vector Machine",classes=class_names)
plot_confusion_matrix(confusion_matrix(y_test, y_pred_dt),"Decision Tree",classes=class_names)
plot_confusion_matrix(confusion_matrix(y_test, y_pred_rf),"Random Forest",classes=class_names)
plot_confusion_matrix(confusion_matrix(y_test, y_pred_knn),"K Nearest Neighbor",classes=class_names)
plot_confusion_matrix(confusion_matrix(y_test, y_pred_adb),"Ada Boost",classes=class_names)
plot_confusion_matrix(confusion_matrix(y_test, y_pred_gb),"Gradient Boosting",classes=class_names)
plt.show()
```
